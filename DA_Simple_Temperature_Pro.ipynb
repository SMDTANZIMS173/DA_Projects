Temperature Analysis
Objective:
Analyze temperature data to find patterns, such as the hottest and coldest days, and visualize trends over time.

```python
import numpy as np
import pandas as pd
from datetime import datetime,timedelta
start_date = datetime(2024,12,21)
dates = [start_date + timedelta(days = i) for i in range(31)]
temperature = [np.random.randint(15,35) for i in range(31)]

# Create a data frame
data = pd.DataFrame({"Dates":dates,"Temperature":temperature})
data.to_csv("Temp_data.csv",index = False)

# Lets apply analysis on this data
#1. Maximum and minimum temperatures
print(f"The Maximum Temperature of 31 days is {data["Temperature"].max()}°C on {data[data["Temperature"] == data["Temperature"].max()]["Dates"].iloc[0].date()}")
print(f"The Minimum Temperature of 31 days is {data["Temperature"].min()}°C on {data[data["Temperature"] == data["Temperature"].min()]["Dates"].iloc[0].date()}")

#2. The average temperature of 31 days is
print(f"The average temperature of 31 days is {data['Temperature'].mean().round(2)}°C")

#3 Days with temperature above 30deg
hot_days = data[data["Temperature"] > 30]
print("Days with temperature above 30deg:")
print(hot_days)
```

```python
# lets plot the temperature trend
import matplotlib.pyplot as plt
plt.figure()
plt.plot(data["Dates"],data["Temperature"],marker = "o",markersize=10,linestyle = "--",linewidth = 1,color = "blue")
plt.title("Temperature Analysis")
plt.xlabel("Dates")
plt.ylabel("Temperature (°C)")
plt.xticks(data["Dates"][::5],rotation = 45)
plt.grid()
plt.show()
```
